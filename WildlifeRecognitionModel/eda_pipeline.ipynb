# Exploratory Data Analysis on the Results of the Cascaded Pipeline

The purpose of this notebook is to evaluate the results of the pipeline. The three metrics used for evaluation are as follows:

1. **The Drop Rate Per Species:** Calculated by $$ \frac{Dropped Images Per X Species}{Total Images Per X Species}$$ This metric evaluates if YOLO has a structural bias against the more rare species of the Mexican Gray Wolf.

2. **The Representation Bias in Processing:** This metric evaluates the class distribution of the dropped files.

3. **The Yield Metric:** This metric determines the total valid biological data points is the pipeline retaining per class under a static threshold of 0.25
